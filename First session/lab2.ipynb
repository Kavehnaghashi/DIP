```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('./pics/lena_gray_256.tif',0)#       
# window_name = 'img'
plt.imshow(image, cmap="gray") #       
plt.axis('off') #      
plt.show()#  
```

```python
histogram = cv2.calcHist([image],[0],None,[256],[0,256])#                        256          0  256    
plt.plot(histogram, color='r') #               
plt.show()#   
```

```python
#img = cv2.imread('./lena_color_512.tif')
#img = cv2.imread('./lena_color_512.tif', cv2.IMREAD_COLOR)
#rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#plt.imshow(rgb)
#plt.axis('off')
#plt.show()

from PIL import Image
pil = Image.open('./pics/lena_color_512.tif') #    
img = np.array(pil)  # shape: (H,W,3) or (H,W) etc. #        

# If grayscale, convert to RGB so matplotlib shows color (if needed)
if img.ndim == 2: #     
    img = np.stack([img]*3, axis=-1)#           

plt.imshow(img)#    
plt.axis('off')#      
plt.show()#   
```

```python
#img = plt.imread("./pics/bird.jpg")
img = cv2.imread('./pics/bird.jpg')#      bgr  
plt.imshow(img)#     rgb  
plt.axis('off')#     
plt.show()#                   

for i,col in enumerate(['b','g','r']): #                0  1  2   
    hist = cv2.calcHist([img],[i],None,[256],[0,256]) #             256    0  256  
    plt.plot(hist, color=col)#             
    plt.xlim([0,256])#     0  256  
    plt.show()
```

```python
gray_img =  cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #           
gray_img_eqhist =  cv2.equalizeHist(gray_img) #                    
```

```python
plt.imshow(gray_img, cmap="gray") #            
plt.axis('off') #      
plt.show()#   
```

```python
# plt.plot(gray_img_eqhist, color='k')
# plt.show()

plt.imshow(gray_img_eqhist,cmap = 'gray') #         
plt.show()#   
```

```python
ret = cv2.threshold(gray_img,80,255,cv2.THRESH_BINARY) #    80   
```

```python
flip = cv2.flip(img,0) #            0  
plt.imshow(flip) #        
plt.axis('off') #      
plt.show()#       
```

```python
blurImg = cv2.blur(img,(20,20)) #         
plt.imshow(blurImg)#   
plt.axis('off')#     
plt.show()#  
```

```python
gausBlur = cv2.GaussianBlur(img,(33,33),0) #        
plt.imshow(gausBlur)#  
plt.axis('off')#     
plt.show()#  
```

```python
medBlur = cv2.medianBlur(img,15)#                  
plt.imshow(medBlur)#  
plt.axis('off')#     
plt.show()#   
```
