```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('./pics/bird.jpg') #     bgr   
plt.imshow(image) #     rgb   
plt.axis('off')#     
plt.show()#                  
```

```python
cropped_image = image[80:180,150:390] #      
plt.imshow(cropped_image)#      
plt.axis('off') #    
plt.show() #   
```

```python
resize_image = cv2.resize(image, [250,250]) #      
plt.imshow(resize_image)#     
plt.axis('off')#    
plt.show() #  
```

```python
resize_image = cv2.resize(image, [500,500], interpolation=cv2.INTER_NEAREST) #         interpolation   zooming
plt.imshow(resize_image)#     
plt.axis('off')#    
plt.show()#  
```

```python
resize_image = cv2.resize(image, [500,500], interpolation=cv2.INTER_LINEAR) #             #  
plt.imshow(resize_image)#     
plt.axis('off')#    
plt.show()#  
```

```python
resize_image = cv2.resize(image, [500,500], interpolation=cv2.INTER_CUBIC)#      16                    
plt.imshow(resize_image)#     
plt.axis('off')#    
plt.show()#  
```

```python
rotate_image = cv2.rotate(image, rotateCode=cv2.ROTATE_90_CLOCKWISE) #   90      
plt.imshow(rotate_image)#    
plt.axis('off')#    
plt.show()#  
```

```python
rotate_image = cv2.rotate(image, rotateCode=cv2.ROTATE_180)#      180  
plt.imshow(rotate_image)#    
plt.axis('off')#    
plt.show()#  
```

```python
rotate_image = cv2.rotate(image, rotateCode=cv2.ROTATE_90_COUNTERCLOCKWISE)#     90     
plt.imshow(rotate_image)#    
plt.axis('off')#    
plt.show()#  
```
