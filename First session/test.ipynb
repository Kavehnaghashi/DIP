```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from sys import getsizeof
```

```python
img = plt.imread("./test/cameraman.tif")
print("shape of image:", img.shape)
print("type of elems of image:", img.dtype)
print("size of image:", getsizeof(img) / 1024 / 1024, "MB")
```

```python
img2 = plt.imread("./test/cats.jpg")
print("shape of image:", img2.shape)
print("type of elems of image:", img2.dtype)
print("size of image:", getsizeof(img2) / 1024 / 1024, "MB")
```

```python
img3 = plt.imread("./test/cats2.jpg")
print("shape of image:", img3.shape)
print("type of elems of image:", img3.dtype)
print("size of image:", getsizeof(img3) / 1024 / 1024, "MB")
```

```python
image = cv2.imread('./test/cats2.jpg',0)
plt.imshow(image, cmap="gray") 
plt.axis('off') 
plt.show()
```

```python
histogram = cv2.calcHist([img2],[0],None,[256],[0,256])
plt.plot(histogram, color='r')
plt.show()
```

```python
resize_image = cv2.resize(img3, [5000,5000], interpolation=cv2.INTER_NEAREST) 
plt.imshow(resize_image)
plt.axis('off')
plt.show()
```

```python
resize_image = cv2.resize(img3, [5000,5000], interpolation=cv2.INTER_CUBIC) 
plt.imshow(resize_image)
plt.axis('off')
plt.show()
```

```python
rotate_image = cv2.rotate(img, rotateCode=cv2.ROTATE_90_CLOCKWISE) 
plt.imshow(rotate_image,cmap='gray')
plt.axis('off')
plt.show()
```

```python
rotate_image = cv2.rotate(img, rotateCode=cv2.ROTATE_90_COUNTERCLOCKWISE) 
plt.imshow(rotate_image,cmap='gray')
plt.axis('off')
plt.show()
```

```python
image2 = cv2.imread("./test/cats.jpg")
image2_copy = image2.copy()
circle_image = cv2.circle(image2_copy, (50,53), 30, (55,55,255), 3) 
circle_image_rgb = cv2.cvtColor(circle_image, cv2.COLOR_BGR2RGB)
plt.imshow(circle_image_rgb) 
circle_image2 = cv2.circle(image2_copy, (123,43), 30, (55,255,255), 3) 
circle_image_rgb = cv2.cvtColor(circle_image2, cv2.COLOR_BGR2RGB)
plt.imshow(circle_image_rgb) 
circle_image3 = cv2.circle(image2_copy, (219,84), 30, (255,55,55), 3) 
circle_image_rgb = cv2.cvtColor(circle_image3, cv2.COLOR_BGR2RGB)
plt.imshow(circle_image_rgb)
plt.axis('off')
plt.show()
```
