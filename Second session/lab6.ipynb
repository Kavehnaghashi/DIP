```python
import cv2
import matplotlib.pyplot as plt 
import math
import numpy as np
```

```python
img = cv2.imread("./pics/cameraman.tif", cv2.IMREAD_GRAYSCALE)

print("Bit depth:", img.dtype)

bit_planes = []

for bit in range(8):
    plane = (img >> bit) & 1     #       
    plane = plane * (2 ** bit)   #              
    bit_planes.append(plane.astype(np.uint8)) #           

# Display
plt.figure(figsize=(10,6)) #    10  6  
for i in range(8):
    plt.subplot(2,4,i+1) #                    
    plt.imshow(bit_planes[i], cmap='gray') #            
    plt.title(f'Bit {i}')#       
    plt.axis('off')#      

plt.tight_layout() #           
plt.show() #     

```

```python
img = cv2.imread("./pics/lena_color_512.tif")   # BGR format


# Split into channels
B, G, R = cv2.split(img) #      

channels = [B, G, R] #   
names = ["Blue", "Green", "Red"]#   

plt.figure(figsize=(12, 12)) #   12  12  

plot_index = 1 #   

# Process each channel
for ch, name in zip(channels, names): #           
    for bit in range(8):
        bit_plane = ((ch >> bit) & 1) * (2**bit) #              
        plt.subplot(3, 8, plot_index) #           
        plt.imshow(bit_plane, cmap='gray') #            
        plt.title(f"{name} Bit {bit}") #        
        plt.axis('off') #       
        plot_index += 1 #       

plt.tight_layout() #           
plt.show() #     
```
